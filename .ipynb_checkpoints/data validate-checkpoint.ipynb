```python
i=dataset[[x for x in dataset.columns if "solv_access" in x]].dropna().index

x=np.array(list(dataset[[x for x in dataset.columns if "solv_access" in x]].loc[i].mean(1)))
y=np.array(dataset["psisolv"].loc[i])

scipy.stats.linregress(x,y).rvalue**2
# r value is .425

# model = LinearRegression()
# model.fit(x, y)
```
