```python
def tri_chips(n_segments):
    # gives triangles (up and down) in one half of the matrix
    
    # as far as I'm concerned this is black magic
    # mess with it at your own risk
    # -KO

    tri=[]

    segx=list(range(0,l,math.floor(l/n_segments)))+[l]
    segy=list(range(0,l,math.floor(l/n_segments)))+[l]

    # downwards triangles
    for j in range(0,n_segments):
        for i in range(j,n_segments):
            tri.append(
                [(y,x) for y in range(segy[j],segy[j+1]) for x in range(y-segy[j]+segx[i],segx[i+1])]
            )

    # upwards triangles
    for j in range(0,n_segments-1):
        for i in range(j+1,n_segments):
            tri.append(
                [(y,x) for y in range(segy[j],segy[j+1]) for x in range(segx[i],y-segy[j+1]+segx[i+1])]
            )
            
    return tri
```

```python
def tri_chips2(n_segments):
    # give only downward triangles in all of the matrix
    # since the bottom half down triangles contain the same info as the top half up triangles
    # also this means all the triangles are the same shape and don't have to be rotated
    
    
    # as far as I'm concerned this is black magic
    # mess with it at your own risk
    # -KO

    tri=[]

    segx=list(range(0,l,math.floor(l/n_segments)))+[l]
    segy=list(range(0,l,math.floor(l/n_segments)))+[l]

    # downwards triangles
    for j in range(0,n_segments):
        for i in range(0,n_segments):
            tri.append(
                [(y,x) for y in range(segy[j],segy[j+1]) for x in range(y-segy[j]+segx[i],segx[i+1])]
            )
            
    return tri
```

```python
# we've eyeballed a chip size of 25 amino acids in length
# we'll need to be able to convolve it across the space.  

chip_size=25
step_size=10


# similar to before, but instead of splitting half the matrix up into non-overlapping chips
# we implement a rolling window which scans across half the space
def convolve_tri_chip(chip_size, step_size, aa_length):
    tri=[]
    for bigy in range(0,aa_length-chip_size,step_size):
        for bigx in range(1,aa_length-bigy-chip_size,step_size):
            tri.append(
                [(y,x) for y in range(bigy,bigy+chip_size) for x in range(y+bigx,bigy+bigx+chip_size)]
            )
    return tri

# this calls convolve_tri_chip
# and saves the chips in a proper csv filfe
def chips_to_csv(protein_id, chip_size, step_size):
    
    protein = np.array(distance_maps_cb['1hzfA0'])
    aa_length=len(protein[0])
    
    chips=convolve_tri_chip(chip_size, step_size, aa_length)
    
    cols=["protein_id","chip_id"]+["px_"+str(n) for n in range(0,len(chips[0]))]

    chiplist=[]
    for i in range(0,len(chips)):
        row=[protein_id,"chip_"+str(i)]
        for n in range(0,len(chips[i])):
            row.append(
                protein[chips[i][n][0],chips[i][n][1]]
            )
        chiplist.append(row)

    chip_df=pandas.DataFrame(chiplist,columns=cols)

    return chip_df
```

```python
# new chipping strategy
# 10 aa (to start with)
# 20 aa
# step size of 1
# 
# go down the diagonal instead of across
# only grab chips for local protein info
# 
# start with 10 proteins
# 
# include other info, maybe as an average for 1d stuff
# and 2d info within triangle
# also ground truth via api
```

```python
#                 [(y,x) for y in range(bigy,bigy+chip_size) for x in range(y+bigx,bigy+bigx+chip_size)]


chip_size=10
step_size=1

aa_length=256

def chip_diagonal(chip_size, step_size, aa_length):
    tri=[]
    for diag in range(1,aa_length-chip_size,step_size):
        tri.append(
            [(y,x) for y in range(diag,diag+chip_size) for x in range(y,chip_size+diag)]
        )
        # NB: y already has diag added to it
    return tri


def gif_from_tri(protein_id,tri,filename):
    
    protein = np.array(distance_maps_cb[protein_id])

    gif=[]

    for t in tri:
        new_test=np.zeros((l,l))
        for (y,x) in t:
            new_test[y,x]=test[y,x]

        im=plt.imshow(new_test)

        x=im.make_image("AGG")[0]
        x=np.flipud(x)

        gif.append(np.array(x))

    imageio.mimsave(filename+'.gif', gif, fps=5)

def diag_chips_to_df(protein_id, chip_size, step_size):
    
    protein = np.array(distance_maps_cb[protein_id])
    aa_length=len(protein[0])
    
    chips=chip_diagonal(chip_size, step_size, aa_length)
    
    cols=["protein_id","chip_id"]+["px_"+str(n) for n in range(0,len(chips[0]))]

    chiplist=[]
    for i in range(0,len(chips)):
        row=[protein_id,"chip_"+str(i)]
        for n in range(0,len(chips[i])):
            row.append(
                protein[chips[i][n][0],chips[i][n][1]]
            )
        chiplist.append(row)

    chip_df=pandas.DataFrame(chiplist,columns=cols)

    return chip_df
```
